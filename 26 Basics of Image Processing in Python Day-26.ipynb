## Basics of Image Processing in Python

### Problem Statement
<p>Guess what is this image? </p>

```python
# import libraries
from matplotlib import pyplot as plt
from skimage import data
from skimage.feature import blob_dog, blob_log, blob_doh
from skimage.color import rgb2gray
import glob
from skimage.io import imread
import math
```

```python
# read image
file = imread('sky.gif')
plt.imshow(file, cmap='gray')
plt.show()
```

```python
# Find the number of Stars in the image
# find blobs
blobs_log = blob_log(file, max_sigma=30, num_sigma=10, threshold=.1)
# Compute radii in the 3rd column.
blobs_log[:, 2] = blobs_log[:, 2] * math.sqrt(2)
numrows = len(blobs_log)
print('Number of Stars: ', numrows)
```

```python
fig, ax = plt.subplots(1, 1)
plt.imshow(file, cmap='gray')
for blob in blobs_log:
    y, x, r = blob
    c = plt.Circle((x, y), r, color='lime', linewidth=2, fill=False)
    ax.add_patch(c)
plt.show()
```
